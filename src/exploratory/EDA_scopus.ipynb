```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel(r"./../../data/raw/scopus-14juli25-00212226.xlsx", engine="openpyxl")
```

```python
df['Year'] = pd.to_numeric(df['Year'], errors='coerce')
tahun_count = df['Year'].value_counts().sort_index()

plt.figure(figsize=(14, 6))
sns.barplot(x=tahun_count.index, y=tahun_count.values)
plt.title("Jumlah Judul Penelitian per Tahun")
plt.xlabel("Tahun")
plt.ylabel("Jumlah Judul")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
